```python
import networkx as nx
import matplotlib.pyplot as plt

def create_directed_graph(array):
    n = len(array)
    G = nx.DiGraph()

    # Add vertices to the graph
    for vertex, _ in array:
        G.add_node(vertex)

    # Add edges based on the positions
    for i, (vertex, position) in enumerate(array):
        left_neighbor_index = (i - position) % n
        right_neighbor_index = (i + position) % n

        left_neighbor = array[left_neighbor_index][0]
        right_neighbor = array[right_neighbor_index][0]

        G.add_edge(vertex, left_neighbor)
        G.add_edge(vertex, right_neighbor)

    return G

def draw_graph(G):
    pos = nx.circular_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=2000, node_color="skyblue", font_size=12, font_weight="bold", arrowsize=20)
    plt.title("Directed Graph")
    plt.show()

def main():
    # Example input array
    array = [('I', 2), ('A', 5), ('E', 4), ('F', 1), ('T', 2), ('S', 3)]

    # Create directed graph
    G = create_directed_graph(array)

    # Draw the graph
    draw_graph(G)
main()

```
